**<center><h2>Results</h2></center>**


In 2012, there were 136,980 deaths registered as suicides or accidental deaths in India, a 20.8% increase from 2001. This equates to a 6.7% increase from 2001 when taking population growth into account. The number of deaths by suicide or accident varies greatly between States/Union Territories (UT's). In terms of the total number of deaths, Tamil Nadu, West Bengal, Andhra Pradesh, and Karnataka had the highest total deaths, together accounting for 44% of all deaths by suicide or accident in India in 2012. 

In terms of the death rate, or the number of deaths by suicide or accident per 100,000 of the population, the mean death rate in 2012 for Indian States/UT’s was 13.5. However, death rates vary greatly between States/UT's. Puducherry experienced the highest death rate in 2012 with 43.4 deaths per 100,000 of the population, followed by the Andaman and Nicobar Islands (31.8), Sikkim (29.6) and Kerala (25.4). There is a fairly clear trend of States/UT's in the north of India having relatively low death rates, and States in the South higher death rates. Sikkim has an unusually high death rate for its geographic location. 

The majority (71%) of States/UT's experienced an increase in death rates between 2001 and 2012. Mizoram experienced the largest increase of 159%, followed by Jammu and Kashmir (122%), while Nagaland experienced the greatest decline in death rates, dropping by 25%. In 2012, Union Territories had a mean death rate 27% higher than States, though this was skewed by two Union Territories with unusually high death rates, namely Puducherry (43.4) and the Andaman and Nicobar Islands (31.8).


The death rate for all age groups follows a relatively stable upward trend from 2001 to 2012. Those between the ages of 30-59 experienced consistently higher death rates than other age groups from 2001 to 2012, while those 14 years and younger consistently had the lowest death rate. The death rate for those 14 years and younger declined by 14% between 2001 and 2012, while for those 30-59, the death rate increase by 10%. The greatest change over time was for those 60 years and older, for whom death rates increased by 12% between 2001 and 2012.

In 2012, males were almost twice as likely to die from suicide or accidental death than females. This gap increased over time: while death rates for females declined by 9% from 2001 to 2012, death rate for males increased by 13%. Death by suicide or accident also tended to be more gender neutral below the age of 29 with males accounting for 55% of all deaths, while for older age brackets (30+), males composed 74% of all deaths. 

While males dominate all occupational categories in terms of number of deaths by suicide or accident, housewife, which was 100% female, was the second largest category of total number of deaths, preceded only by a general 'other' occupational category. 

There is a strong positive relationship between the population size of States/UTs and number of deaths by suicide or accident, with population size explaining 79% of the variance in number of deaths by suicide or accident. The best fit model was of a cubic polynomial form. However, this model excluded two of the 35 observations for which this model did not fit, and this was not due to faulty data. These two observations were Uttar Pradesh and Bihar, which have very large populations and very low deaths by accident or suicide. The reason(s) why Uttar Pradesh and Bihar have such a low number of deaths by suicide or accident while having such large populations should be investigated in further research.


**<h3>1.5 Handling incorrect data </h3>**


Exploratory data analysis indicated issues with West Bengal's data. West Bengal recorded 39,660 deaths in 2011, while in 2012 there were 0. Data for all other States and UT's in 2012 follow the trends of previous years. Therefore, it is most likely that 2012 data for West Bengal is incorrect (missing). As the data for all other States and UT's follow a logical trend for 2012, 2011 data was used as estimates of suicides and deaths for West Bengal in 2012. 

```python
# Take rows of 2011 and 2012 data for WEST BENGAL
temp_2011 = df[(df['STATE/UT'] == 'WEST BENGAL') & (df['Year'] == 2011)]
temp_2012 = df[(df['STATE/UT'] == 'WEST BENGAL') & (df['Year'] == 2012)]
# Keep a record of indices of rows of 2012 data (in order to delete them from the main dataframe later, and use them to modify temp_2011)
ind_2012 = temp_2012.index
# Delete all 2012 data for WEST BENGAL from the main dataframe
df = df.drop(ind_2012, axis = 0)
# Modify column `Year` in temp_2011 (so that they can be added to the main dataframe)
temp_2011['Year'] = 2012
# Replace the index of temp_2011 with those of temp_2012
temp_2011.index = ind_2012
# Add the modified temp_2011 to the main dataframe
df = pd.concat([df, temp_2011], axis = 0)
# Sort rows by index (because temp_2011 was added to the end of df)
df = df.sort_index()

# Total rows are no longer valid so are dropped
index_names = df[(df['STATE/UT'] == 'TOTAL (STATES)') | (df['STATE/UT'] == 'TOTAL (UTs)') | (df['STATE/UT'] == 'TOTAL (ALL INDIA)')].index
df.drop(index_names, inplace = True)
```

In addition, Ladakh is missing from the data set, likely treated as part of Jammu and Kashmir. Thus, the eight territories that will form the basis of this analysis include:

1. The Andaman and Nicobar Islands
2. Chandigarh
3. Dadra and Nagar Haveli 
4. Daman and Diu
5. Delhi
6. Jammu and Kashmir
7. Lakshadweep
8. Puducherry

In 2012, UT's composed 2% of the total population of India, with a population of 32,390050. The majority (90%) of the population of UT's are situated in two territories: Delhi (16,787941, 52%) and Jammu and Kashmir (12,267032, 37.9%).

In 2012, States had a Mean death rate of 12.4, while UT's had a Mean death rate of 17, 27% higher that States. **<h3>2.5.3 Testing for homoscedasticity </h3>**



```python
x = no_outliers['Pop_Total_2012']
y = no_outliers['Total_deaths_by_State']
x = sm.add_constant(x)
model = sm.OLS(y, x)
model_result = model.fit()
test = sms.het_breuschpagan(model_result.resid, model_result.model.exog)
names = ['Lagrange multiplier statistic', 'p-value',
        'f-value', 'f p-value']
lzip(names, test)
```

The Breusch-Pagan Test was used to determine whether heteroscedasticity is present in the model. The result was statistically significant (p = .005) with a Lagrange multiplier of 7.6. Thus, the null hypothesis of homoscedasticity is rejected, and a transformation is required. Thus, number of deaths is transformed through the log transformation, which is often used to spread out clumped data, and bring together spread-out data.

```python
Deaths_log = np.log(no_outliers['Total_deaths_by_State'])
x = no_outliers['Pop_Total_2012']
y = Deaths_log
x = sm.add_constant(x)
model = sm.OLS(y, x)
model_result = model.fit()
test = sms.het_breuschpagan(model_result.resid, model_result.model.exog)
names = ['Lagrange multiplier statistic', 'p-value',
        'f-value', 'f p-value']
lzip(names, test)
```

The Breusch-Pagan Test of the Log transformed number of deaths and population size was not statistically significant (p = .45), indicating that the null hypothesis can be accepted that the data is homoscedastic.

**<h3>2.5.6 Findings </h3>**


For the vast majority of States/UT's, population size explains 80% of the variance in the number of deaths by suicide or accident. This makes theoretical sense as the more people there are, the more people there are who could die of an accident or suicide. However, this model did not apply to Uttar Pradesh and Bihar, which have very large populations and very low deaths by accident or suicide. The reason(s) why these States have such low number of deaths while having such large populations should be investigated in further research.
